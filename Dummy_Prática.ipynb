```python
import pandas as pd
import numpy as np
import seaborn as sns
import statsmodels.formula.api as smf
```

```python
df = pd.read_excel('limite_cr√©dito_escolaridade.xlsx')
df.head()
```

```python
df = pd.get_dummies(df, prefix=['Escolaridade'])
df.head()
```

```python
df.corr()
```

```python
df.columns
```

```python
function = 'LimitedoChequeEspecial~Idade+RendimentoTotal+Salario+LimitedeCreditoImediato+Escolaridade_Fundamental_Medio+Escolaridade_Superior_Pos'
model = smf.ols(formula=function, data=df).fit()
print(model.summary())
```
